```python
# 1. Import required library
import pandas as pd

# 2. Create student performance dataset
data = {
    'attendance': [90, 60, 80, 50, 95, 70],
    'assignment': [85, 55, 78, 45, 90, 65],
    'quiz': [88, 50, 75, 40, 92, 60],
    'final_result': ['Pass', 'Fail', 'Pass', 'Fail', 'Pass', 'Fail']
}

# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# 3. Display the dataset
print("Student Performance Dataset:")
print(df)

# 4. Identify input features and target variable
X = df[['attendance', 'assignment', 'quiz']]
y = df['final_result']

print("\nInput Features:")
print(X)

print("\nTarget Variable:")
print(y)

# 5. Identify one pattern from the dataset
print("\nIdentified Pattern:")
print("Students with higher attendance, assignment, and quiz scores tend to PASS,")
print("while students with lower scores tend to FAIL.")
```
